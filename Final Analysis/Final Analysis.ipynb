```python
import pandas as pd
```

```python
df =pd.read_csv('cleaned_books.csv')
df
```

```python
# Add a column to calculate discount prices by applying a 10% discount on each book price.
discount_price = df['Price']*0.10
df['10% Discount'] = discount_price
df.to_csv('updated_file.csv', index=False)
```

```python
# What's the average price of all books?
avg_price = df[['Price']].mean()
print(f'The average price of the books is {avg_price}')
```

```python
# How many books are "In Stock" vs. "Out of Stock"?
in_stock_count = df[df['Availability'].str.contains("In stock", na=False)].shape[0]
out_of_stock_count = df[df['Availability'].str.contains("Out of Stock", na=False)].shape[0]
print(f"In Stock: {in_stock_count}")
print(f"Out of Stock: {out_of_stock_count}")
```
